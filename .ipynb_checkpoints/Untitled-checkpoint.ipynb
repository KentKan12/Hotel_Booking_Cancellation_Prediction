# 🏨 Prediksi Pembatalan Pemesanan Hotel (Hotel Booking Cancellation Prediction)

## 📖 Latar Belakang
Pembatalan pesanan hotel (*cancellation*) berdampak signifikan terhadap pendapatan dan efisiensi operasional hotel. Kamar yang dibatalkan mendadak seringkali sulit disewakan kembali, menyebabkan potensi kehilangan pendapatan (*revenue loss*). Di sisi lain, *overbooking* sebagai langkah antisipasi yang tidak terukur bisa menurunkan kepuasan pelanggan. Oleh karena itu, diperlukan sebuah sistem analitik prediktif cerdas untuk memperkirakan probabilitas pembatalan sejak awal tamu melakukan pendaftaran (*booking*).

## 🎯 Tujuan
Membangun model *Machine Learning* yang mampu memprediksi apakah seorang tamu akan membatalkan pesanannya atau tidak, **tepat pada saat mereka selesai melakukan proses *booking*** di sistem.

## 💼 Manfaat
- **Optimalisasi Pendapatan**: Memungkinkan pihak manajemen hotel untuk menerapkan kebijakan *overbooking* yang lebih terukur, aman, dan berbasis data.
- **Efisiensi Operasional**: Membantu dalam mengalokasikan staf dan merencanakan kebutuhan logistik (persediaan makanan, penyiapan kamar) dengan lebih akurat.
- **Strategi Pemasaran & Kebijakan**: Mengidentifikasi segmen pelanggan yang berisiko tinggi batal sehingga hotel dapat memberlakukan kebijakan deposit khusus, penyesuaian harga, atau menahan inventaris kamar untuk pelanggan yang lebih pasti.

## 💡 Insight Utama yang Didapatkan
1. **Paradoks Non-Refund**: Tamu dengan tipe deposit *Non-Refund* justru memiliki tingkat pembatalan tertinggi (hampir 100% di segmen tertentu). Hal ini disinyalir sebagai aktivitas *fraud* (pemesanan kamar sementara untuk syarat pengajuan visa) dari luar negeri, bukan perilaku tamu normal.
2. **Golden Rule Parkir**: Tamu yang memesan tempat parkir (berapapun jumlah mobilnya) memiliki tingkat pembatalan **0%**. Ini adalah sinyal terkuat dari niat kedatangan tamu yang sesungguhnya.
3. **Riwayat Pembatalan**: Lebih dari 90% tamu yang pernah membatalkan pesanan di masa lalu cenderung akan membatalkan kembali pesanannya saat ini.
4. **Segmen & Distribusi Rawan**: Pemesanan rombongan (Groups) yang dilakukan melalui agen perjalanan (TA/TO) memiliki risiko batal ekstrim (>65%), membutuhkan pengawasan dan kebijakan blokir kalender yang lebih hati-hati.

## ⚙️ Feature Engineering & Model
- **Feature Engineering**: Meliputi ekstraksi risiko geografis (Country Risk), kepadatan agen (Agent Density), deteksi penipuan visa (Visa Fraud Suspect), dan pengelompokan ukuran tamu (Guest Type). Fitur-fitur yang membocorkan data masa depan (*Data Leakage*) seperti `reservation_status`, `assigned_room_type`, dan `booking_changes` dihapus secara agresif.
- **Model yang Digunakan**: Fokus pada algoritma *Tree-Based* dan *Ensemble* modern seperti Random Forest, XGBoost, HistGradientBoosting, LightGBM, dan CatBoost. Pembobotan kelas (*Class Weights/Scale Pos Weight*) diaktifkan untuk menangani *imbalanced data*.

## 📊 Evaluasi & Rekomendasi Model Berdasarkan Kondisi Bisnis
Dalam proyek ini, terjadi persaingan ketat antara 3 konfigurasi model terbaik:
1. **Baseline LightGBM (Threshold 0.50)**: Unggul dalam keakuratan umum dengan **Accuracy 80.06%** dan F1-Score 0.7680, namun kurang agresif dalam menangkap pembatalan (Recall 80.32%).
2. **Threshold HistGradientBoosting (Threshold 0.39)**: Sangat agresif dengan **Recall mencapai 90.78%** (hanya menyisakan sedikit False Negative), namun mengorbankan ketepatan tebakan (Precision turun ke 67.12%).
3. **Threshold LightGBM (Threshold 0.42)**: Menjadi *Sweet Spot* bisnis dengan **F1-Score tertinggi (0.7745)**, mengamankan **Recall tinggi (87.32%)** dengan penurunan Precision yang jauh lebih minim (69.58%) dibandingkan HistGradientBoosting.

**Rekomendasi Utama (Umum):** 
Model **LightGBM dengan Threshold 0.42** dipilih sebagai model terbaik untuk operasional harian hotel. Model ini memberikan proteksi optimal dari hilangnya pendapatan akibat kamar kosong (menangkap 87.32% pembatalan nyata), sekaligus meminimalkan risiko tuntutan hukum dan kompensasi akibat *overbooking* massal (karena mampu menjaga *True Negative* dan menekan *False Positive* dengan lebih stabil).

# SPLITTING DATA

### Pembagian Data (Data Splitting): Mencegah Kebocoran Waktu
Dalam kasus bisnis riil seperti perhotelan, data sangat dipengaruhi oleh waktu dan musim (*seasonality*). Jika kita membagi data Train dan Test secara acak (*Random Split*), model berisiko "mengintip" tren di masa depan (*Data Leakage*), yang membuat akurasinya terlihat tinggi saat *training* namun gagal saat diimplementasikan ke sistem asli. Pendekatan **Time Series Split** dilakukan dengan mengurutkan data berdasarkan waktu kedatangan (Tahun, Bulan, Hari). Data 80% pertama bertindak sebagai histori masa lalu (Train), dan 20% sisanya bertindak sebagai pengujian masa depan (Test).

Gunakan Time Series Split, agar model dapat mempelajari pola booking musiman dan menghindari data leakage

```python
# Mapping nama bulan menjadi angka (1-12)
months_map = {
    'January':1, 'February':2, 'March':3, 'April':4, 'May':5, 'June':6,
    'July':7, 'August':8, 'September':9, 'October':10, 'November':11, 'December':12
}
df['arrival_date_month'] = df['arrival_date_month'].map(months_map)

# Buat datetime object sementara untuk mengambil hari dalam seminggu
temp_date = pd.to_datetime(
    df['arrival_date_year'].astype(str) + '-' + 
    df['arrival_date_month'].astype(str) + '-' + 
    df['arrival_date_day_of_month'].astype(str)
)
df['arrival_day_of_week'] = temp_date.dt.dayofweek

# buat fitur cycle waktu dari bulan untuk mendeteksi pola musiman per bulannya 
df['month_sin'] = np.sin(2 * np.pi * df['arrival_date_month'] / 12.0)


# sort data berdasarkamn tahun, bulan, lalu minggu
df = df.sort_values(by=['arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month']).reset_index(drop=True)
split_index = int(len(df) * 0.8)
train_df = df.iloc[:split_index].copy()
test_df = df.iloc[split_index:].copy()

# Kita buang fitur ini untuk mencecegah multikolinearistas
train_df = train_df.drop(columns=['arrival_date_month'])
test_df = test_df.drop(columns=['arrival_date_month'])
```

# Cleaning Data

### Pembersihan Data (Data Cleaning): Penanganan Nilai Kosong
Penanganan *missing values* tidak sekadar membuang atau mengisi dengan nilai rata-rata, melainkan dilakukan dengan pendekatan logika operasional hotel:
1. **Country**: Jumlahnya sangat kecil (<5%), sehingga baris data aman untuk dihapus tanpa menghilangkan banyak informasi.
2. **Agent & Company**: Data kosong bukan berarti datanya hilang, melainkan merepresentasikan bahwa tamu tersebut memesan secara mandiri (*Direct/Personal*) tanpa perantara agen atau penjaminan perusahaan. Nilai kosong diisi dengan `0` untuk merepresentasikan pola pemesanan langsung tersebut.
3. **Children**: Jika data anak kosong, sangat masuk akal jika diasumsikan bahwa tamu tersebut tidak membawa anak kecil (diisi `0`).

## DELETE INCONSISTENT VALUE IN ADR AND NUM PEOPLE OF STAY

### Penghapusan Data Anomali & Pencegahan Data Leakage
Langkah kritis ini bertujuan untuk memastikan model hanya belajar dari kondisi yang benar-benar diketahui sistem **saat tamu pertama kali mendaftar (booking)**:
- **Data Anomali**: Dihapus karena merusak distribusi pola harga. Harga (ADR) minus tidak masuk akal secara finansial, harga ADR ekstrim (5400) adalah pencilan/kesalahan ketik, dan pesanan tanpa manusia (0 adults, 0 children, 0 babies) mengindikasikan *ghost booking* atau cacat data.
- **Data Leakage (Kebocoran Masa Depan)**: Fitur seperti `reservation_status`, `assigned_room_type` (tipe kamar yang akhirnya diberikan saat tamu tiba), dan `booking_changes` (perubahan setelah booking awal) memuat informasi dari masa depan. Memasukkan fitur ini akan membuat model "berbuat curang" dan gagal berfungsi di lingkungan aslinya. Kolom-kolom ini **wajib dibuang**.

```python
# Hapus data dengan ADR minus (di bawah 0)
train_df = train_df[train_df['adr'] > 0]
test_df = test_df[test_df['adr'] > 0]
# Hapus outlier ekstrem (ADR 5400). Kita batasi ADR maksimal di angka wajar ( maks 1000)
train_df = train_df[train_df['adr'] <= 1000]
test_df = test_df[test_df['adr'] <= 1000]

# Kita hanya menyimpan data yang total tamunya LEBIH DARI 0
train_df = train_df[~((train_df['adults'] == 0) & (train_df['children'] == 0) & (train_df['babies'] == 0))]
test_df = test_df[~((test_df['adults'] == 0) & (test_df['children'] == 0) & (test_df['babies'] == 0))]
```

Nilai adr yang bernilai 0 akan mengacaukan perhitungan rata-rata adr sehingga harus dihapus dari data. Selain itu hanya terdapat 1 outlier yang nilai nya sangat jauh dari data(adr = 5400), Oleh karena itu drop saja data tersebut

## Booking Type Analysis

### Feature Engineering: Analisis Risiko Segmen & Saluran Distribusi
Tidak semua saluran pemesanan diciptakan sama. Melalui pemetaan silang (*Crosstab & Heatmap*), ditemukan adanya kuadran risiko yang sangat jelas. Pemesanan grup melalui agen (TA/TO) memiliki tingkat pembatalan ekstrem (65%). Sementara tamu *Corporate* atau *Direct* sangat jarang membatalkan pesanan (<15%). Insight ini diekstrak menjadi satu klasifikasi fitur kategorikal baru (`segment_channel_risk`) agar model lebih tajam mengenali "zona bahaya" dari sebuah jalur pemesanan.

```python
df_analysis = train_df.copy()

# Hitung Rasio Pembatalan (Mean) untuk setiap kombinasi
pivot_cancel_rate = df_analysis.pivot_table(
    index='market_segment', 
    columns='distribution_channel', 
    values='is_canceled', 
    aggfunc='mean'
).fillna(0) # Mengisi kombinasi yang tidak ada datanya dengan 0

# Hitung Total Volume Pesanan (Count) untuk melihat sebaran data
pivot_booking_count = df_analysis.pivot_table(
    index='market_segment', 
    columns='distribution_channel', 
    values='is_canceled', 
    aggfunc='count'
).fillna(0)
print("MATRIKS RASIO PEMBATALAN")
display(pivot_cancel_rate.style.background_gradient(cmap='Reds', axis=None).format("{:.2%}"))

print("\n MATRIKS VOLUME PESANAN (Melihat seberapa banyak data riilnya)")
display(pivot_booking_count.style.background_gradient(cmap='Blues', axis=None).format("{:,.0f}"))


# VISUALISASI PLOT HEATMAP

plt.figure(figsize=(12, 5))

# Plot Persentase Batal
plt.subplot(1, 2, 1)
sns.heatmap(pivot_cancel_rate, annot=True, fmt=".1%", cmap="OrRd", cbar=False)
plt.title("Rasio Pembatalan per Kombinasi")
plt.ylabel("Market Segment")
plt.xlabel("Distribution Channel")

# Plot Jumlah Pesanan
plt.subplot(1, 2, 2)
sns.heatmap(pivot_booking_count, annot=True, fmt=",.0f", cmap="Blues", cbar=False)
plt.title("Total Volume Pesanan per Kombinasi")
plt.ylabel("")
plt.xlabel("Distribution Channel")

plt.tight_layout()
plt.show()
```

Dari visualisasi kita bisa membagi 4 zona resiko pembatalan booking berdasarkan segmen dan distribusi channel, yaitu:
1. Zona tingkat pembatalan sangat tinggi:
   - Group booking via TA/TO, memiliki tingkat pembatalan 65%
2. Zona tingkat pembatalan tinggi
   - offline TA/TO via corporate, memiliki rasio mencapai 49,52%.
   online atau offline TA via TA/TO, keduanya mencapai rasio pembatalan 34-35%
   , gabungan dari 2 segmen ini lebih dari 60.000 pesanan
   Yang terakhir adalah group via direct dan corporate, sebesar 32-36%
3. Zona resiko sedang
   - coporate via corporate % TA/TO, sebesar 18 %-21 % tingkat pembatalan dan
   aviation via corporate & online TA via GDS beada di rentang 20-23%
4. Zona aman
   - direct via direct sebesar 14% dan online TA via direct & direct via TA/TO
   sebesar dibawah 10% (minim)


```python
# buat fitur baru berdasarkan resiko pembatalan dari segmen dan distribution channel
def assign_segment_channel_risk(row):
    segment = str(row['market_segment'])
    channel = str(row['distribution_channel'])
    combination = f"{segment}_via_{channel}"
    
    # 1. Extreme Risk Zone (>60% Cancel Rate)
    if combination in ['Groups_via_TA/TO']:
        return 'Extreme_Risk'
    
    # 2. High Risk Zone (30% - 50% Cancel Rate)
    elif combination in [
        'Offline TA/TO_via_Corporate', 
        'Groups_via_Direct', 
        'Groups_via_Corporate', 
        'Online TA_via_TA/TO', 
        'Offline TA/TO_via_TA/TO', 
        'Online TA_via_Corporate'
    ]:
        return 'High_Risk'
    
    # 3. Medium Risk Zone (15% - 30% Cancel Rate)
    elif combination in [
        'Aviation_via_Corporate', 
        'Corporate_via_TA/TO', 
        'Online TA_via_GDS', 
        'Corporate_via_Corporate',
        'Offline TA_via_Direct',
        'Offline TA_via_GDS',
        'Complementary_via_Corporate'
    ]:
        return 'Medium_Risk'
        
    # 4. Low Risk Zone (<15% Cancel Rate & Sisa Data Langka)
    else:
        return 'Low_Risk'

for df_split in [train_df, test_df]:
    # Buat fitur risiko baru
    df_split['segment_channel_risk'] = df_split.apply(assign_segment_channel_risk, axis=1)
print('DONE')
```

## Deposit Type Analysis

### Feature Engineering: Mengungkap Paradoks Deposit (Deteksi Penipuan Visa)
Secara logika, deposit *Non-Refundable* (uang tidak kembali) seharusnya menjamin kedatangan tamu. Namun visualisasi data mengungkap anomali: hampir 100% tamu *Non-Refund* di beberapa segmen justru membatalkan pesanan. Ini adalah pola aktivitas sindikat/pemesanan blokir palsu (*fraud*) dari luar negeri untuk sekadar memenuhi dokumen pengajuan Visa. Untuk mengatasi bias aneh ini pada model, dibuatlah fitur intelijen `visa_fraud_suspect` khusus untuk mendeteksi pesanan dengan kombinasi *Non Refund*, berlokasi dari luar negara asal (Portugal), dan melalui jalur rawan.

```python
sns.set_theme(style="whitegrid")

plt.figure(figsize=(14, 7))
ax = sns.barplot(
    x='market_segment', 
    y='is_canceled', 
    hue='deposit_type', 
    data=train_df, 
    errorbar=None,          
    palette='Set2'         
)

# Mengubah format angka di sumbu Y menjadi persentase (0.2 -> 20%)
import matplotlib.ticker as mtick
ax.yaxis.set_major_formatter(mtick.PercentFormatter(1.0))

plt.title('Analisis Paradoks: Tingkat Pembatalan Berdasarkan Segmen Pasar & Tipe Deposit', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Segmen Pasar (Market Segment)', fontsize=12, labelpad=10)
plt.ylabel('Tingkat Pembatalan (Cancellation Rate)', fontsize=12, labelpad=10)
plt.legend(title='Tipe Deposit', loc='upper right', frameon=True)
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

1. Non refund hampir selalu menyentuh angka 100% pada setiap segmen. Ini menunjukkan bahwa non refund bukan lagi kebijakan finansial yang menahan orang untuk batal, tetapi sinyal kuat mengenai adanya ketidakpastian yang sangat tinggi di segmen-segmen tersebut.
2. Jika dilihat, tingkat pembatalan non refund pada  TA/TO dan Groups hampir 100%, ini berarti agen penyelenaggara/perjalanan sering kali melakkan block booking yang berakhir dengan pembatalan massal

3. Berdasarkan makalah dari penerbit datanya, lonjakan pembatalan pada pesanan Non-Refundable bukanlah error sistem, melainkan perilaku riil dari sindikat visa palsu luar negeri. Jika baris ini dihapus, model akan mengalami blind spot (titik buta) dan menganggap semua pesanan Non-Refund di masa depan pasti aman, padahal justru paling rawan batal.

<!-- Untuk fitur ini kita biarkan saja, tidak perlu membuat fitur untuk mendeteksi booking palsu ini, karena model tree bisa menemukan pola keanehan sendiri dari melakukan percabangan pohon -->

untuk mengatasi masalah booking palsu ini, kita akan buat fitur untuk mendeteksi booking palsu tersebut

```python
train_df['is_portugal'] = (train_df['country'] == 'PRT').astype(int)
test_df['is_portugal'] = (test_df['country'] == 'PRT').astype(int)
df_split['booking_pipeline'] = df_split['market_segment'].astype(str) + '_via_' + df_split['distribution_channel'].astype(str)
for df_split in [train_df, test_df]:
    df_split['booking_pipeline'] = df_split['market_segment'].astype(str) + '_via_' + df_split['distribution_channel'].astype(str)
    # Deteksi Visa Fraud yang jauh lebih presisi (Menggunakan konstrain Geografis luar Portugal)
    cond_fraud = (
        (df_split['deposit_type'] == 'Non Refund') & 
        (df_split['is_portugal'] == 0) & 
        (df_split['booking_pipeline'].str.contains('TA/TO|Online TA|Groups'))
    )
    df_split['visa_fraud_suspect'] = cond_fraud.astype(int)
    df_split.drop(columns=['market_segment','distribution_channel','booking_pipeline','is_portugal'], inplace=True)

print('DONE')
```

# ENCODING DATA

### Encoding & Pengecekan Multikolinearitas
Karena algoritma *Machine Learning* hanya mengenali bentuk numerik, nilai-nilai teks/kategorikal dipetakan ke dalam angka (Ordinal). Untuk kategori teks yang cukup rumit seperti ukuran grup tamu (`guest_type_category`), algoritma pengubah **Target Encoding** dimanfaatkan, yang merubah nama kategori menjadi persentase risiko pembatalan historisnya secara otomatis.
Uji VIF (*Variance Inflation Factor*) digunakan untuk membuang redundansi antar fitur (menghindari fitur ganda), sementara *Mutual Information* membuktikan variabel mana yang paling berkorelasi positif dengan pembatalan (*target class*).

```python
target = "is_canceled"
X_train = train_df.drop(columns=[target])
y_train = train_df[target]

X_test = test_df.drop(columns=[target])
y_test = test_df[target]

print("=" * 40)
print("Train Feature :", X_train.shape)
print("Train Target  :", y_train.shape)
print("Test Feature  :", X_test.shape)
print("Test Target   :", y_test.shape)
```

```python
mapping_hotel = {'City Hotel': 1, 'Resort Hotel': 0}
mapping_deposit = {'No Deposit': 0, 'Refundable': 1, 'Non Refund': 2}
mapping_segment = {'Extreme_Risk': 4,'High_Risk': 3, 'Medium_Risk': 2, 'Low_Risk': 1}
mapping_lead = {'long-term': 4,'mid-term': 3, 'short-term': 2, 'last-minute': 1}
mapping_country = {'High_Risk_Country': 3, 'Medium_Risk_Country': 2, 'Low_Risk_Country': 1}

for df in [X_train, X_test]:
    df['deposit_type'] = df['deposit_type'].map(mapping_deposit)
    df['country_risk_bin'] = df['country_risk_bin'].map(mapping_country)
    df['segment_channel_risk'] = df['segment_channel_risk'].map(mapping_segment)
    df['cat_lead_time'] = df['cat_lead_time'].map(mapping_lead)
    df['hotel'] = df['hotel'].map(mapping_hotel)
print('DONE')
```

```python
from category_encoders import TargetEncoder

# Membuat encoder
target_encoder = TargetEncoder(cols='guest_type_category')

# Fit pada data train
X_train = target_encoder.fit_transform(X_train, y_train)

# Transform data test
X_test = target_encoder.transform(X_test)
```

# MODELLING

### Pemodelan: Training & Evaluasi Model Baseline
Sebanyak 9 algoritma kompetitif berbasis Pohon Keputusan (*Tree-Based*) dan *Ensemble* dirancang sebagai baseline. Algoritma kelompok pohon dipilih secara khusus karena sangat tangguh (*robust*) terhadap nilai pencilan (seperti pada Lead Time & Harga/ADR), dan dapat memetakan aturan interaksi non-linear secara alami tanpa standardisasi (*scaling*). Masalah dominasi data kelas 0 (tamu tidak batal) ditangani secara bawaan (internal) pada model lewat injeksi *Class Weights / Scale Pos Weight* (menyeimbangkan prioritas).

```python
import time

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import (
    RandomForestClassifier,
    ExtraTreesClassifier,
    AdaBoostClassifier,
    HistGradientBoostingClassifier
)

from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    classification_report,
    confusion_matrix
)


# 9 MODEL BASELINE Dengan Balancing
majorit_count = sum(y_train == 0) 
minority_count = sum(y_train == 1) 
weight_ratio = majorit_count / minority_count
models = {
    "Decision Tree": DecisionTreeClassifier(
        random_state=42,
        class_weight="balanced"
    ),

    "Random Forest": RandomForestClassifier(
        n_jobs=-1,
        random_state=42,
        class_weight="balanced"
    ),

    "Extra Trees": ExtraTreesClassifier(
        n_jobs=-1,
        random_state=42,
        class_weight="balanced"
    ),

    "AdaBoost": AdaBoostClassifier(
        random_state=42
    ),

    "Hist Gradient Boosting": HistGradientBoostingClassifier(
        random_state=42,
        class_weight="balanced"
    ),

    "LightGBM": LGBMClassifier(
        random_state=42,
        verbose=-1,
        class_weight="balanced"
    ),

    "XGBoost": XGBClassifier(
        use_label_encoder=False,
        eval_metric="logloss",
        random_state=42,
        scale_pos_weight = weight_ratio
    ),

    "CatBoost": CatBoostClassifier(
        verbose=0,
        random_state=42,
        auto_class_weights='Balanced',
    )
}

# Variabel untuk menyimpan 2 model baseline terbaik(bisa disesuaikan dengan hasil training)

lgb_baseline_model = None
hist_baseline_model = None

# Wadah hasil evaluasi
results = []

print("=== MEMULAI ARENA KOMPETISI 9 MODEL BASELINE ===")
print(f"Data Train: {len(X_train):,} baris | Data Test: {len(X_test):,} baris\n")


# Training & Evaluasi
for name, model in models.items():

    start_time = time.time()

    # ---------------------------
    # Training
    # ---------------------------
    model.fit(X_train, y_train)

    # Simpan model baseline
    if name == "LightGBM":
        lgb_baseline_model = model

    elif name == "Hist Gradient Boosting":
        hist_baseline_model = model

    # Prediksi
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

 
    # Classification Report
    print("\n" + "="*70)
    print(f"MODEL : {name}")
    print("="*70)

    print(classification_report(
        y_test,
        y_pred,
        digits=4
    ))


    # TP, TN, FP, FN
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()

    print("Detail Prediksi")
    print(f"TP : {tp:,}")
    print(f"TN : {tn:,}")
    print(f"FP : {fp:,}")
    print(f"FN : {fn:,}")

    # Evaluasi
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_prob)

    run_time = time.time() - start_time

    results.append({
        "Model": name,
        "Accuracy": acc,
        "Precision": prec,
        "Recall": rec,
        "F1-Score": f1,
        "ROC-AUC": roc_auc,
        "Time (s)": run_time
    })

    print(f"\n✔️ [{name}] Selesai dilatih dalam {run_time:.2f} detik.")


# Hasil Akhir
results_df = pd.DataFrame(results).sort_values(
    by="ROC-AUC",
    ascending=False
)

print("\n")
print("="*90)
print("HASIL AKHIR KOMPETISI BASELINE MODEL")
print("="*90)

print(results_df.to_string(
    index=False,
    formatters={
        "Accuracy": "{:.4f}".format,
        "Precision": "{:.4f}".format,
        "Recall": "{:.4f}".format,
        "F1-Score": "{:.4f}".format,
        "ROC-AUC": "{:.4f}".format,
        "Time (s)": "{:.2f}".format
    }
))


# Model yang tersimpan
print("\nModel baseline berhasil disimpan:")
```

```python
from sklearn.metrics import precision_recall_curve, auc, average_precision_score

y_prob_baseline = lgb_baseline_model.predict_proba(X_test)[:, 1]

precision, recall, thresholds = precision_recall_curve(y_test, y_prob_baseline)

pr_auc_score = auc(recall, precision)
avg_precision = average_precision_score(y_test, y_prob_baseline)

print("========================================================")
print("             EVALUASI KETIDAKSEIMBANGAN DATA            ")
print("========================================================")
print(f"Skor PR-AUC (Luas Area)  : {pr_auc_score:.4f}")
print(f"Average Precision (AP)   : {avg_precision:.4f}")
print("========================================================")

plt.figure(figsize=(8, 6))

plt.plot(
    recall,
    precision,
    color="darkorange",
    lw=2,
    label=f"LightGBM (PR-AUC = {pr_auc_score:.4f})"
)

baseline = y_test.mean()

plt.axhline(
    y=baseline,
    color="navy",
    linestyle="--",
    label=f"Baseline Tanpa Model (Rasio Batal = {baseline:.4f})"
)

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("Recall (Kemampuan Menangkap Tamu Batal)")
plt.ylabel("Precision (Ketepatan Tebakan Batal)")
plt.title("Kurva Precision-Recall (PR Curve) - LightGBM Baseline")
plt.legend(loc="lower left")
plt.grid(True, linestyle="--", alpha=0.5)

plt.show()
```

```python
import time
from sklearn.model_selection import TimeSeriesSplit, RandomizedSearchCV
from sklearn.metrics import classification_report, f1_score, roc_auc_score


# 1. TIME SERIES SPLIT
tscv = TimeSeriesSplit(n_splits=5)

# 2. PARAMETER GRID
# Hist Gradient Boosting
param_grid_hgb = {
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'max_iter': [100, 150, 200],
    'max_leaf_nodes': [15, 31, 63, 127],
    'min_samples_leaf': [20, 50, 100],
    'l2_regularization': [0.0, 0.1, 1.0, 10.0]
}

# LightGBM
param_grid_lgb = {
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'n_estimators': [100, 150, 200],
    'num_leaves': [15, 31, 63, 127],
    'max_depth': [-1, 5, 10, 15],
    'min_child_samples': [20, 50, 100],
    'reg_alpha': [0.0, 0.1, 1.0],
    'reg_lambda': [0.0, 0.1, 1.0]
}

# 3. WADAH MODEL TERBAIK
best_models = {}


# 4. HIST GRADIENT BOOSTING TUNING
print("=== START TUNING: HIST GRADIENT BOOSTING ===")

start_hgb = time.time()

hgb_search = RandomizedSearchCV(
    estimator=hist_baseline_model,
    param_distributions=param_grid_hgb,
    n_iter=15,
    scoring="f1",
    cv=tscv,
    n_jobs=1,
    random_state=42
)
hgb_search.fit(X_train, y_train)

best_models["Hist Gradient Boosting"] = hgb_search.best_estimator_

print(f" HGB Tuning Selesai, Waktu: {time.time() - start_hgb:.1f} detik")
print("Parameter Terbaik HGB:")
print(hgb_search.best_params_)
print()


# 5. LIGHTGBM TUNING
print("=== START TUNING: LIGHTGBM ===")

start_lgb = time.time()

lgb_search = RandomizedSearchCV(
    estimator=lgb_baseline_model,
    param_distributions=param_grid_lgb,
    n_iter=15,
    scoring="f1",
    cv=tscv,
    n_jobs=1,
    random_state=42
)

lgb_search.fit(X_train, y_train)

best_models["LightGBM"] = lgb_search.best_estimator_

print(f"LightGBM Tuning Selesai, Waktu: {time.time() - start_lgb:.1f} detik")
print("Parameter Terbaik LightGBM:")
print(lgb_search.best_params_)
print()
```

```python
print("EVALUASI FINAL MODEL HASIL TUNING PADA DATA TEST")

# Loop untuk menguji setiap model terbaik yang disimpan di dictionary best_models
for name, model in best_models.items():
    print(f"\n MODEL: {name}")
    print("-" * 50)
    
    # 1. Prediksi label (0 atau 1)
    y_pred = model.predict(X_test)
    
    # 2. Prediksi probabilitas untuk kelas 1 (Canceled) untuk hitung ROC-AUC
    y_prob = model.predict_proba(X_test)[:, 1]
    
    # 3. Tampilkan Classification Report lengkap
    print("[Classification Report]")
    print(classification_report(y_test, y_pred, digits=4))
    
    # 4. Tampilkan Skor ROC-AUC
    roc_auc = roc_auc_score(y_test, y_prob)
    print(f"ROC-AUC Score: {roc_auc:.4f}")
    print("-" * 50)
```

# THRESHOLD SETTING

## HIST BOOSTING

```python


# 1. AMBIL PROBABILITAS DATA TEST DARI MODEL BASELINE TERBAIK

y_prob_baseline = hist_baseline_model.predict_proba(X_test)[:, 1]

# 2. ITERASI UNTUK MENCARI THRESHOLD TERBAIK BERDASARKAN F1-SCORE
thresholds = np.arange(0.01, 1.0, 0.01)
f1_scores = []

for t in thresholds:
    # Jika probabilitas >= t, maka beri label 1, selain itu 0
    y_pred_dynamic = (y_prob_baseline >= t).astype(int)
    f1_scores.append(f1_score(y_test, y_pred_dynamic))

# 3. TEMUKAN SKOR TERTINGGI
best_index = np.argmax(f1_scores)
best_threshold = thresholds[best_index]
best_f1_score = f1_scores[best_index]


print(f"Threshold Optimal : {best_threshold:.2f}")
print(f"F1-Score Maksimal : {best_f1_score:.4f}\n")


y_pred_final = (y_prob_baseline >= best_threshold).astype(int)

print(f"=== CLASSIFICATION REPORT DENGAN THRESHOLD {best_threshold:.2f} ===")
print(classification_report(y_test, y_pred_final, digits=4))
```

## LGB

```python
y_prob_baseline = lgb_baseline_model.predict_proba(X_test)[:, 1] 


thresholds = np.arange(0.01, 1.0, 0.01)
f1_scores = []

for t in thresholds:

    y_pred_dynamic = (y_prob_baseline >= t).astype(int)
    f1_scores.append(f1_score(y_test, y_pred_dynamic))

best_index = np.argmax(f1_scores)
best_threshold = thresholds[best_index]
best_f1_score = f1_scores[best_index]

print(f"Threshold Optimal : {best_threshold:.2f}")
print(f"F1-Score Maksimal : {best_f1_score:.4f}\n")

y_pred_final = (y_prob_baseline >= best_threshold).astype(int)

print(f"=== CLASSIFICATION REPORT DENGAN THRESHOLD {best_threshold:.2f} ===")
print(classification_report(y_test, y_pred_final, digits=4))
```

### 💼 Panduan Implementasi Berdasarkan Kondisi Bisnis Hotel

Manajemen tidak harus kaku menggunakan satu model selamanya. Threshold dan model dapat diganti secara dinamis mengikuti siklus musim kunjungan hotel (*Hotel Seasonality*):

#### 🟩 REKOMENDASI 1: Model Terbaik untuk Kondisi Umum & Harian
👉 **PILIHAN: LightGBM dengan Threshold 0.42**
* **Alasan Bisnis:** Model ini memenangkan skor **F1-Score tertinggi (0.7745)**. Secara bisnis, model ini menaikkan kemampuan mendeteksi orang batal dari 80% (pada baseline) menjadi **87.32%** (hanya kecolongan sedikit False Negative). Yang paling krusial, model ini mampu menahan laju pembengkakan *False Positive* (Precision tertahan di 69.58%, lebih baik daripada HistGradientBoosting yang anjlok ke 67%). Ini adalah opsi paling aman untuk menjaga revenue kamar sekaligus melindungi reputasi hotel dari bencana *overbooking* harian.

#### 🟥 REKOMENDASI 2: Model Khusus Periode High-Season (Lebaran, Natal, Tahun Baru)
👉 **PILIHAN: HistGradientBoosting dengan Threshold 0.39**
* **Alasan Bisnis:** Pada musim liburan besar, permintaan kamar sangat tinggi (*over-demand*). Menjual kembali kamar yang kosong akibat pembatalan mendadak sangatlah mudah karena antrean tamu *walk-in* atau pemesan *last-minute* melimpah. Pada kondisi ini, kita ingin **Recall setinggi mungkin (90.78%)** agar tidak ada satu pun pembatalan yang lolos dari deteksi model. Risiko *False Positive* (tamu diprediksi batal tapi ternyata datang) tidak terlalu berbahaya karena hotel bisa dengan mudah memindahkan tamu atau menawarkan *upgrade* kamar premium yang kosong.

#### 🟨 REKOMENDASI 3: Model Khusus Kondisi Low-Season (Hari Kerja / Bulan Sepi)
👉 **PILIHAN: Baseline LightGBM dengan Threshold 0.50**
* **Alasan Bisnis:** Ketika tingkat keterisian hotel sedang sepi, setiap tamu yang datang sangatlah berharga. Kita tidak boleh membuat tamu kecewa akibat kesalahan *overbooking*. Oleh karena itu, kita membutuhkan **Precision tertinggi (73.58%)** untuk memastikan bahwa jika model menebak seseorang akan batal, tebakan itu benar-benar akurat dan terpercaya.

### 🚀 Rekomendasi Tindakan Operasional (Actionable Items)
Bagi hotel yang mengintegrasikan model **LightGBM (Threshold 0.42)** ini ke sistem PMS (*Property Management System*), berikut langkah mitigasi otomatis yang bisa dijalankan:
1. **Pre-emptive Contact:** Jika pemesanan terdeteksi berisiko batal tinggi oleh model, sistem hotel secara otomatis mengirimkan email/WhatsApp konfirmasi H-3 untuk meminta *re-assurance* dari tamu.
2. **Automated Waitlist System:** Kamar dengan probabilitas batal > 42% dapat dibuka kembali di sistem OTA (Online Travel Agent) sebagai slot *Waitlist* berdiskon, sehingga jika terjadi pembatalan sesungguhnya, kamar langsung terisi secara instan tanpa kehilangan ADR (*Average Daily Rate*).
3. **Pengetatan Aturan Deposit Khusus:** Untuk pemesanan luar negeri yang terdeteksi sebagai `visa_fraud_suspect` dengan skor prediksi tinggi, hotel wajib mengubah syarat pembayaran dari *Pay at Hotel* menjadi *Advance Payment* 100% dalam waktu 24 jam untuk memotong ekosistem booking palsu.
